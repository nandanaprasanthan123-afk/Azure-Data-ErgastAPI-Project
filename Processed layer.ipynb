```python
# ADLS access permission
storage_account = "storage_account_name"
access_key = "storage_account_key"

spark.conf.set(
    f"fs.azure.account.key.{storage_account}.dfs.core.windows.net",
    access_key
)
```

```python
# container check
dbutils.fs.ls(
    "abfss://processed@driversdatastorage.dfs.core.windows.net/"
)
```

```python
# read from delta
processed_df = spark.read.format("delta") \
.load("abfss://processed@driversdatastorage.dfs.core.windows.net/drivers")

display(processed_df)
```

```python
# Reading json data from raw container
# raww_df = spark.read.json(
#     "abfss://raww@driversdatastorage.dfs.core.windows.net/drivers_api"
# )

# display(raww_df)
```

```python
# converting json data into delta format
# # saving to presentation layer
processed_df.write \
.format("delta") \
.mode("overwrite") \
.save(
    "abfss://presentation@driversdatastorage.dfs.core.windows.net/drivers"
)
```

```python
# verifying data 
presentation_df = spark.read \
.format("delta") \
.load(
    "abfss://presentation@driversdatastorage.dfs.core.windows.net/drivers"
)

display(presentation_df)
```
